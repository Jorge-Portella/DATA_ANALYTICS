### Limpieza de Nulos y CEROS.

```python

df2 = df[[column for column in df if df[column].count() / len(df) >= 0.2]]
del df2['Id']
print("List of dropped columns:", end=" ")
for c in df.columns:
    if c not in df2.columns:
        print(c, end=", ")
print('\n')
df = df2
```

### Correlation

```python
df_num_corr = df_num.corr()['SalePrice'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

####  Los valores se enceuntran correlacionados, pero se procede a eliminar los valores atípicos para analizar cuál puede tener una buena correlación sin sus valores atípicos.

```python
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['SalePrice'])
```

#### se analiza que la ultima grafica es perfecta ya que se compara SalePrice contra SalePrice. Vamos a eliminar CEROS.

```python
import operator

individual_features_df = []
for i in range(0, len(df_num.columns) - 1): # -1 because the last column is SalePrice
    tmpDf = df_num[[df_num.columns[i], 'SalePrice']]
    tmpDf = tmpDf[tmpDf[df_num.columns[i]] != 0]
    individual_features_df.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['SalePrice'][0] for feature in individual_features_df}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```

```python
#### aparece otra variable furtemente correlacionada golden_features_list
```

```python
golden_features_list = [key for key, value in all_correlations if abs(value) >= 0.5]
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

### Conclusiones

#### Existe una furertecorrelación entreslo valores numéricos, Se tinen 11 características con fuerte relación con SalePrice.  SalePrice muestran un patrón particular con líneas verticales que significan que son variables discretas con un corto rango.

### Relación característica a característica.

```python
corr = df_num.drop('SalePrice', axis=1).corr() # We already examined SalePrice correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

#### La mayoria de las características parecen estar correlacionadas entre sí, pero algunas de ellas, como YearBuild / GarageYrBlt, pueden indicar una inflación de precios a lo largo de los años. En cuanto a 1stFlrSF / TotalBsmtSF, es normal que cuanto más grande sea el primer piso (considerando que muchas casas tienen solo 1 piso), más grande será el sótano total. Ahora, podemos ver que: 

1. Existe una fuerte correlación negativa entre BsmtUnfSF (pies cuadrados sin terminar de área del sótano) y BsmtFinSF2 (pies cuadrados terminados tipo 2). 

2. HalfBath / 2ndFlrSF puede indicar que la gente le da una importancia a no tener que bajar corriendo las escaleras en caso de tener que ir al baño urgentemente.

3. Podemos concluir que algunas de esas funciones pueden combinarse entre sí para reducir la cantidad de funciones (1stFlrSF / TotalBsmtSF, GarageCars / GarageArea)
